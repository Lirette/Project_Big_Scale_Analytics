# V- Deployement on the cloud
### 1- Deployement on App Engine 
*  Creation of a file named requirement.txt containing all usefull dependencies to run the application

*   And finally we've depolyed  it on App Engine by using the the  gcloud command ***gcloud app deploy***. It tooks several minutes but in the end we received the link of our web application

### 2- Deployement on Cloud Run
*  Update  lingorank_app.py accordingly
*  Creation of a Dockerfile
*  Build a docker image and deploy the conterized application with the gcloud commands: ***gcloud builds submit --tag gcr.io/french-text-difficulty-350020/predict*** and ***gcloud run deploy --image gcr.io/french-text-difficulty-350020/predict --platform managed*** where predict is the containers' name. Minutes later the link of the dockerised flask app was available.
